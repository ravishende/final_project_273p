```python
test_cols = [col for col in df.columns if "test/" in col]
relevant_cols = ["Name"] + test_cols
relevant_cols
```

```python
def model_name(name:str) -> str:
    """get a nicer name of the model"""
    rename = {
        "resnet18-real_artifact_net-aggregate-final":"ResNet18 Hybrid",
        "resnet18-rgb-aggregate-final":"ResNet18 RGB",
        "resnet18-fft-aggregate-final":"ResNet18 FFT"
    }
    return rename[name]

def get_results(df, test:bool=True) -> df:
    datasets = ['cifake', 'hemg', 'rajarshi']
    metrics = ['acc', 'auc', 'f1', 'loss', 'precision', 'recall']
    metric_titles = {
        "acc": "Accuracy",
        "auc": "AUC",
        "f1": "F1",
        "loss":"Loss",
        "precision":"Precision",
        "recall":"Recall"
    }
    # Compute average across datasets for each metric
    results_df = pd.DataFrame(columns=['Name'] + metrics)
    results_df["Name"] = df["Name"]
    split = "test" if test else "train"
    for m in metrics:
        cols = [f'test/{d}/{m}' for d in datasets]
        results_df[f'{m}'] = df[cols].mean(axis=1)
    results_df = results_df.rename(columns=metric_titles)
    results_df = results_df.dropna(axis=0, how='any')
    results_df["Name"] = results_df["Name"].apply(model_name)
    return results_df
```

```python
test_results_df = get_results(df, test=True)
train_results_df = get_results(df, test=False)
```
